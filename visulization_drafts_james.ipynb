```python
import pandas as pd
import altair as alt
```

```python
df_5 = pd.read_excel('SAYO_T_AltForm.xlsx')
df_5.head()
```

```python
domain = ["Pre", "Post"]
range_ = ["#7570b3", "#d95f02"]

chart = alt.Chart(df_5).mark_bar().encode(
    x=alt.X('Assessment:N', sort='y'),
    y='mean(Score):Q', 
    color = alt.Color("Assessment", scale=alt.Scale(domain=domain, range=range_)),
    column='Category:N'
).properties(
    width=115,
    height=300,
    title={"text": "Average Total Score Improvements of Students in BtC '22 from SAYO-T",}
)

chart
```
